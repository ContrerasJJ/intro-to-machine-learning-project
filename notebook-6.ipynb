# Megaline Tariff Recommendation: Smart vs Ultra

## Project Overview

Mobile Carrier **Megaline** discovered that many subscribers are still using legacy plans. The company wants to recommend one of its newer plans - **Smart** or **Ultra** - based subscriber behavior.

In this project, I build a **Classification model** that predicts which plan is best for based on thier monthly activity. The model uses hi
storical data from users who have already switched to one of the new plans.

**Goal:**
Develop a model that predicts the plan ('Smart' or Ultra') with an **accuracy of at least 0.75** on the test set.

**Metric**
- Primary metric: **Accuracy**

**Data source:** 'datasets/users_behavor.csv'

Each row contains monthly behavior of one user:
- 'calls' - number of calls
- 'minutes' - total call duration in minutes
- 'messages' - number of text messages
- 'mb_used' - internet traffic used in MB
- 'is_ultra' - current plan (Ultra = 1, Smart = 0)

```python
# Select features and target

features = ['calls', 'minutes', 'messages', 'mb_used']
target = 'is_ultra'

X = df[features]
Y = df[target]
```

```python
# First split: train+validation vs test (20% test)

X_temp, X_test, Y_temp, Y_test = train_test_split(
    X,Y, test_size=0.2, random_state=12345, stratify=Y
)

# Second split: from remaining 80%, create train (60%) and validation (20%)
# 0.25 of 80% = 20%

X_train, X_valid, Y_train, Y_valid = train_test_split(
    X_temp, Y_temp, test_size=0.25, random_state=12345, stratify=Y_temp
)

X_train.shape, X_valid.shape, X_test.shape
```

## Train / Validation / Test Split

The data was split into three parts:

**Training set:** 60%
**Validation set:** 20%
**Test set:** 20%

Procedure:
1. First, I split the full dataset into:
  - 80% -> temporary set ('X_temp', 'Y_temp')
  - 20% -> test set ('X_test', 'Y_test')
2. Then, I split the temporary set (80%) again into:
  - 75% -> training set (which is 60% of the original data)
  - 25% -> validation set (which is 20% of the original data)

The 'stratify' argument was used in both splits to preserve the class distribution of 'is_ultra' across all sets. This helps to make sure that each subset is representative of the whole dataset.

```python
dummy = DummyClassifier(strategy='most_frequent', random_state=12345)
dummy.fit(X_train, Y_train)

y_valid_dummy = dummy.predict(X_valid)
dummy_valid_accuracy = accuracy_score(Y_valid, y_valid_dummy)
dummy_valid_accuracy
```

## Baseline Model (DummyClassifier)

As a baseline, I used 'DummyClassifier' with the strategy '"most_frequent"', which always predicts the most common class in the training set.

**Validation accuracy of dummy model:** shown in the cell above.

This baseline reflects the performance of a very simple model that does not use any information from the features. Any meaningful model should perform **better** than this baseline.

```python
best_tree_model = None
best_tree_accuracy = 0
best_tree_params = None

for depth in range(1,21):
    model = DecisionTreeClassifier(
        max_depth=depth,
        random_state=12345
    )
    model.fit(X_train, Y_train)

    Y_valid_pred = model.predict(X_valid)
    acc = accuracy_score(Y_valid, Y_valid_pred)

    print(f"DecisionTree: max_depth={depth} -> valid accuracy = {acc:.4f}")

    if acc > best_tree_accuracy:
        best_tree_accuracy = acc
        best_tree_model = model
        best_tree_params = {'max_depth': depth}

print("\nBest Decision Tree params:", best_tree_params)
print("Best Decision Tree validation accuracy:", best_tree_accuracy)
```

```python
best_forest_model = None
best_forest_accuracy = 0
best_forest_params = None

for n_estimators in [10, 30, 50, 100]:
    for depth in [3, 5, 7, 10, None]:
        model = RandomForestClassifier(
            n_estimators=n_estimators,
            max_depth=depth,
            random_state=12345
        )
        model.fit(X_train, Y_train)

        Y_valid_pred = model.predict(X_valid)
        acc = accuracy_score(Y_valid, Y_valid_pred)

        print(f"RandomForest: n_estimators{n_estimators},"
             f"max_depth={depth} -> valid accuracy = {acc:.4f}")

        if acc > best_forest_accuracy:
            best_forest_accuracy = acc
            best_forest_model = model
            best_forest_params = {
                'n_estimators': n_estimators,
                'max_depth': depth
            }

print("\nBest Random Forest params:", best_forest_params)
print("Best Random Forest validation accuracy:", best_forest_accuracy)
```

```python
log_reg = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000, random_state=12345)
)

log_reg.fit(X_train, Y_train)


Y_valid_lr = log_reg.predict(X_valid)
lr_valid_accuracy = accuracy_score(Y_valid, Y_valid_lr)
lr_valid_accuracy
```

```python
results = pd.DataFrame({
    'Model': ['Dummy' + '(most frequent)', 'Decision Tree (best)', 'Random Forest (best)', 'Logistic Regression'],
    'Validation accuracy': [
        dummy_valid_accuracy,
        best_tree_accuracy,
        best_forest_accuracy,
        lr_valid_accuracy
    ]
})

results
```

## Model Comparison and Findings

I evaluated serveral models on the validation set:

**Baseline:** DummyClassifier (most frequent class)
**Decision Tree:** tuned by 'max_depth' from 1 to 20
**Random Forest:** tuned by 'n_estimators' and 'max_depth'
**Logustic Regression:** with standardized features

**Key Observations:**

- The **DummyClassifier** achieved an accuracy around the majority class proportion, which serves as a lower baseline.
- The **Decision Tree** model improved upon the baseline and reached its best performance at around 'max_depth = (0.81)'.
- The **Random Forest** model provided the highest validation accuracy overall. The best hyperparameters found were:
  'n_estimators = (50)'
  'max_depth = (7)'
- **Logistic Regression** also performed reasonably well but did not surpass the best Random Forest configuration.

Based on these results, I selected the **Random Forest** with the best hyperparameters as the final model.

## Sanity Check of the Model

To better understand how the model makes decisions, I performed a simple sanity check:

1. **Class balance vs accuracy**
   - The baseline model (always predicting the majority class) achieved an accuracy close to the majjoity class share.
   - The final Random Forest model significantly outperformed this baseline, confirming that it is learning meaningful patterns rather than just memorizing the most frequent class.

2. **Feature importance**
   - The feature importance scores from the Random Forest model showed that:
     'mb_used' and 'minutes' are typically the most important features.
     'messages' and 'calls' also contribute but have slightly lower importance.
   - This makes sense: internet usage and call duration are good indicators of whether a user needs the more expensize Ultra plan or the cheaper Smart plan.

3. **Error analysis**
  - Some misclassified users have moderate usage levels, where their behavior could reasonably fit either plan.
  - This suggests that borderline cases are naturally harder to classify and that the model is not simply making random errors.

Overall, the models behavior is **reasonable and consistent** with expectations about telecom usage: heavier users are more likely to be classified into the Ultra plan, while lighter users tend to be assigned the Smart plan.
